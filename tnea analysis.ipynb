```python
current_year = 2023
for t in ['C','R']:
    for i in range(1,6):
        resp = rq.get("https://cutoff.tneaonline.org/api/auth/glist/{}{}".format(i,t))
        df=pd.DataFrame(resp.json())
        df.to_csv("data/{}_{}.csv".format(current_year-i,t))
```

```python
current_year = 2023
for t in ['C','R']:
    for i in range(1,6):
        df=pd.read_csv("data/{}_{}.csv".format(current_year-i,t))
        df["year"] = current_year-i
        df.to_csv("data/{}_1.csv".format(t))
```

```python
current_year = 2023
for t in ['C', 'R']:
    # Initialize an empty dataframe
    df = pd.DataFrame()
    for i in range(1,6):
        temp_df = pd.read_csv("data/{}_{}_1.csv".format(current_year-i,t))
        df = df.append(temp_df, ignore_index=True)
    df.to_csv("data/{}_{}_all.csv".format(current_year-i,t))
```
